# 1396. Set Union
#### https://www.lintcode.com/problem/set-union

# Approach 1: Union Find

```python
from collections import defaultdict
class Solution:
    """
    @param sets: Initial set list
    @return: The final number of sets
    """
    def setUnion(self, sets):
        # Write your code here
        n = len(sets)
        self.father = [i for i in range(n)]
        self.size = n
        
        num_to_idxs = defaultdict(list)
        for idx in range(len(sets)):
            for num in sets[idx]:
                num_to_idxs[num].append(idx)
        for num in num_to_idxs:
            root = num_to_idxs[num][0]
            for idx in num_to_idxs[num][1:]:
                self.union(root, idx)

        return self.size
        
    def union(self, a, b):
        A = self.find(a)
        B = self.find(b)
        if A != B:
            self.father[A] = self.find(B)
            self.size -= 1
        
    def find(self, a):
        x = a
        while self.father[x] != x:
            x = self.father[x]
        while self.father[a] != a:
            a, self.father[a] = self.father[a], x
        return x
```
