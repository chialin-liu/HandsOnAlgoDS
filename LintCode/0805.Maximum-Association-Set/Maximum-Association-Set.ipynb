# 805. Maximum Association Set

# Approach 1: Union Find

```python
class Solution:
    """
    @param ListA: The relation between ListB's books
    @param ListB: The relation between ListA's books
    @return: The answer
    """
    def maximumAssociationSet(self, ListA, ListB):
        # Write your code here
        self.father = {}
        self.size = {}
        self.maxsize = 1
        self.maxroot = ""
        n = len(ListA)
        for i in range(n):
            self.insert(ListA[i])
            self.insert(ListB[i])
            self.union(ListA[i], ListB[i])
        books = set(ListA + ListB)
        ans = []
        print(self.maxsize)
        for book in books:
            if self.size[self.find(book)] == self.maxsize:
                ans.append(book)
        
        return ans
    
    def insert(self, a):
        if a not in self.father:
            self.father[a] = a
            self.size[a] = 1
    def union(self, a, b):
        A = self.find(a)
        B = self.find(b)
        if A != B: 
            self.father[A] = B
            self.size[B] += self.size[A]
            self.maxsize = max(self.maxsize, self.size[B])
    def find(self, a):
        x = a
        while self.father[x] != x:
            x = self.father[x]
        while self.father[a] != x:
            a, self.father[a] = self.father[a],x
        return x
```
