# 586. Sqrt(x) II
#### https://www.lintcode.com/problem/sqrtx-ii

# Approach 1: Binary Search

```python
class Solution:
    """
    @param x: a double
    @return: the square root of x
    """
    def sqrt(self, x):
        # write your code here
        left, right = 0, x
        if 0 < right < 1: right = 1 / right
        while left <= right:
            mid = (left + right) / 2.0
            if mid * mid == x: return mid
            elif mid * mid < x:
                left = mid + 0.0000000001
            else:
                right = mid- 0.0000000001
        return right
```

# Approach 2: Newton's Method 

```python
class Solution:
    """
    @param x: a double
    @return: the square root of x
    """
    def sqrt(self, x):
        result = 1.0;  
        while abs(result * result - x) > 1e-12:
            result = (result + x / result) / 2
        
        return result       
```
